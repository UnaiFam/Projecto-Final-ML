```python
import requests
def get_zip_info(zip_code):

    try:
        url = f"https://api.zippopotam.us/us/{int(zip_code)}"
        response = requests.get(url)
        if response.status_code == 200:
            return us_states[response.json()["places"][0]["state"]]
        else:
            return "Unknown"
    except Exception as e:
        return "Unknown"
    

```
