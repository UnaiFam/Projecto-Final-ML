```python
X=df.drop(columns=["Consumer disputed?"])
```

```python
y=df['Consumer disputed?'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
y
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras import layers

# --- Asumiendo que ya tienes el DataFrame df cargado y limpiado de NaNs bÃ¡sicos ---

# 1. CodificaciÃ³n binaria de la variable objetivo


from imblearn.over_sampling import SMOTE



X_encoded = pd.get_dummies(X, drop_first=True)

# --- 4. Aplicar SMOTE (balanceo de clases) ---
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_encoded, y)



# 5. Separar X (features) e y (target)


X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.20, random_state = 24)
```

```python
model = tf.keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Salida binaria
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 7. Entrenamiento del modelo
model.fit(X_train, y_train, 
          epochs=15, 
          batch_size=32, 
          validation_split=0.1, 
          verbose=1)

# 8. EvaluaciÃ³n en test
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```
