```python
X=df_processed.drop(columns=["Consumer disputed?"])
```

```python
y=df_processed["Consumer disputed?"]
```

```python
from imblearn.over_sampling import ADASYN
```

```python
X_resampled, y_resampled = ADASYN().fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size = 0.20, random_state = 24)
```

```python
from sklearn.svm import SVC
```

```python
import optuna
```

```python
from sklearn.metrics import accuracy_score
```

```python
def objective(trial):
    param = {

 "C": trial.suggest_categorical("C", [0.00001, 0.5, 1]),
    "kernel":trial.suggest_categorical ("kernel", ['linear']),


}
    

    svc_class1 = SVC(**param, random_state=24, probability=True)
    svc_class1.fit(X_train, y_train)

    pred = svc_class1.predict(X_test)
    return accuracy_score(y_test, pred)
```

```python
study = optuna.create_study(
    storage="sqlite:///db.sqlite3",  # Almacenamiento
    study_name="SVC DISPUTE",
    direction='maximize',
    sampler=optuna.samplers.TPESampler(), #CmaEsSampler(), # Probad con otros
    load_if_exists=True
)
```

```python
from sklearn.decomposition  import PCA
from sklearn.pipeline import make_pipeline
```

```python
svc_class_def = make_pipeline(

    PCA(n_components=20),  # Retiene 95% de la varianza
    SVC(random_state= 24, probability=True)
)
```

```python

svc_class_def.fit(X_train, y_train)

```

```python
y_proba = svc_class_def.predict_proba(X_test)[:, 1]
y_pred = (y_proba > 0.5).astype(int)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = (svc_class_def.predict_proba(X_test)[:, 1] >= 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred, normalize='true'))
```

```python
from sklearn.model_selection import cross_validate
cross_val=cross_validate(svc_class_def, X_train, y_train, cv=10)["test_score"]

print(cross_val.mean())
print(cross_val.std())
```
