```python
from sklearn.base import ClassifierMixin, BaseEstimator

class MiClasificador(ClassifierMixin, BaseEstimator):
    def __sklearn_tags__(self):
        # AsegÃºrate que la clase padre tenga este mÃ©todo
        tags = super().__sklearn_tags__()
        # AquÃ­ puedes modificar los tags si quieres
        return tags
```

```python



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 24)
```

```python
categorical_features=["Product", "Sub-product" , "Issue",   
"Sub-issue",         
"State", "Company response",  "Timely response?"]


```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features), 

    ],

    remainder='passthrough'  # deja pasar las numÃ©ricas sin tocar
)
```

```python
n_features=preprocessor.fit_transform(X_train).shape[1]
n_features
```

```python
def build_model ():
    model = tf.keras.Sequential([
        keras.layers.Input(shape=(n_features,)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')  # Salida binaria
    ])

    model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy'])

    return model
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("callback_model.keras")
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10)
```

```python
keras_clf = KerasClassifier(
                            model=lambda: build_model(),
                            epochs=50, callbacks=[checkpoint_cb, early_stopping_cb ], validation_split=0.2, batch_size=100, verbose=0)
```

```python
model_neural = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('Classifier', keras_clf)  ])
```
